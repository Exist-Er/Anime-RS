```python
# imports
%load_ext cudf.pandas

import pandas as pd
```

```python
# reading dataset

anime = pd.read_csv("../data/Anime_processed.csv")
user = pd.read_csv('../data/users/user_1001133.csv')
```

```python
rating = user.set_index('anime_id', drop=True)['rating']

matrix_1 = anime.set_index('anime_id', drop=True)
matrix_2 = pd.merge(user.drop(['rating'], axis=1), anime, on='anime_id', how='inner').set_index('anime_id', drop=True).T

CBF_score = matrix_1.dot(matrix_2)
```

```python
x = CBF_score
x['CBF_score'] = x.mul(rating, axis=1).sum(axis=1, ascending=False)    # x.sum(axis=1) was not workig with cudf
```

```python
x['CBF_score'].sort_values(ascending=False).to_csv('../data/temp.csv')
```
