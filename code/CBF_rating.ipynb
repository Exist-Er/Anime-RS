```python
# imports

import pandas as pd
```

```python
# reading dataset

anime = pd.read_csv("../data/Anime_processed.csv")
```

```python
user = pd.read_csv('../data/users/user_1001133.csv')
rating = user.set_index('anime_id', drop=True)

matrix_1 = anime.set_index('anime_id', drop=True)
matrix_2 = pd.merge(user.drop(['rating'], axis=1), anime, on='anime_id', how='inner').set_index('anime_id', drop=True).T

print(matrix_1.shape)
print(matrix_2.shape)

CBF_score = (matrix_1 @ matrix_2)

print(CBF_score)
```

```python
CBF_score = CBF_score * rating.T
print(CBF_score)
```

```python
def calculate_CBF_score(file: str) -> None:
    user = pd.read_csv(file)
    rating = user['rating']

    matrix_1 = anime.set_index('anime_id', drop=True)
    matrix_2 = pd.merge(user, anime, on='anime_id', how='inner').set_index('anime_id', drop=True).T


    CBF_score = (matrix_1 @ matrix_2)
```
