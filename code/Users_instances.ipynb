```python
# include

import pandas as pd
```

```python
# reading dataset

users = pd.read_csv("../archive/users-score-2023.csv")
```

```python
# minimum number of records for a user file
MIN_LEN = 400

for user_id, user_ratings in users.groupby('user_id'):
    if user_ratings.shape[0] > MIN_LEN:
        user_df = user_ratings[['anime_id', 'rating']]
        user_df.to_csv(f'../data/users/user_{user_id}.csv', index=False)
```
